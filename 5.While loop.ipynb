# While loop

#Python While Loop is used to execute a block of statements repeatedly until a given condition is satisfied. And when the condition becomes false, the line immediately after the loop in the program is executed.


# write a program to print the following using while loop
#a- first 10 even numbers
#b- first 10 odd numbers
#c- first 10 natural numbers
#d- first 10 whole numbers

```python
num=2
while num<=20:
    print("EVEN NUMBER",num)
    num=num+2
```

```python
num=1
while num<=20:
    print("ODD NUMBER",num)
    num=num+2
```

```python
num=1
while num<=20:
    print("NATURAL NUMBER",num)
    num=num+1
```

```python
num=0
while num<10:
    print("WHOLE NUMBER",num)
    num=num+1
```

# write a program to print first 10 integers and theirs squares using while loop 

```python
num=1
print("Numbers \t\t Squares")
while num<=10000:
    print(num,"\t\t\t",num**2)
    num=num+1
```

# write a program to print first 10 natural number in reverse order using while loop

```python
num=10
while num>=1:
    print(num)
    num=num-1
```

# write a program to print sum of first 10 natural numbers

```python
num=10
sum=0
while num>=1:
    sum= sum+num
    num=num-2
    print(sum)
```

# write a program to print table of a number entered from the user


```python
i=1
num=int(input("Enter any Number"))
while i<=10:
    print(num, "*" ,i, "=" , num*i )
    i=i+1
```
